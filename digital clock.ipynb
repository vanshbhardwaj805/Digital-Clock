```python
import tkinter as tk
import time

def update_clock():
    current_time = time.strftime("%H:%M:%S")  # Get the current time in HH:MM:SS format
    clock_label.config(text=current_time)  # Update the label with the current time
    clock_label.after(1000, update_clock)  # Call this function again after 1000 ms (1 second)

# Create the main window
root = tk.Tk()
root.title("Digital Clock")

# Create a label to display the time
clock_label = tk.Label(root, font=("Helvetica", 48), bg="black", fg="white")
clock_label.pack(pady=20)

# Start the clock update function
update_clock()

# Run the application
root.mainloop()
```
